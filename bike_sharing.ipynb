````python
df = pd.DataFrame(df)
display(df)
````

````python
df= df.drop_duplicates()
df= df.dropna()
display(df)
````

````python
df= df.replace([float('NaN')],'-')
````

# b. Qual a m√©dia da coluna windspeed?


````python
media = df['windspeed'].mean()
print("A m√©dia da coluna √©:", media)
````

# c. Qual a m√©dia da coluna temp?



````python
media = df['temp'].mean()
print("A m√©dia da coluna √©:", media)
````

# d. Quantos registros existem para o ano de 2011?


````python
df['datetime'] = pd.to_datetime(df['datetime'])
df['year'] = df['datetime'].dt.year
registros_2011 = len(df[df['year'] == 2011])
print("O n√∫mero de registros para o ano de 2011 √©:", registros_2011)
````

# e. Quantos registros existem para o ano de 2012?


````python
df['datetime'] = pd.to_datetime(df['datetime'])
df['year'] = df['datetime'].dt.year
registros_2011 = len(df[df['year'] == 2012])
print("O n√∫mero de registros para o ano de 2012 √©:", registros_2012)
````

#f. Quantas loca√ß√µes de bicicletas foram efetuadas em 2011?


````python
locacoes_2011 = df[df['year'] == 2011]['total_count'].sum()
print("Foram efetuadas", locacoes_2011, "loca√ß√µes de bicicletas em 2011")
````

# g. Qual esta√ß√£o do ano cont√©m a maior m√©dia de loca√ß√µes de bicicletas?

````python
grupo_por_estacao = df.groupby('season')
media_por_estacao = grupo_por_estacao['total_count'].mean()
estacao_mais_alta = media_por_estacao.idxmax()
print("Esta√ß√£o com a maior m√©dia de loca√ß√µes de bicicletas:", estacao_mais_alta)
````

# h. Qual esta√ß√£o do ano cont√©m a menor m√©dia de loca√ß√µes de bicicletas?

````python
grupo_por_estacao = df.groupby('season')
media_por_estacao = grupo_por_estacao['total_count'].mean()
estacao_mais_baixa = media_por_estacao.idxmin()
print("Esta√ß√£o com a menor m√©dia de loca√ß√µes de bicicletas:", estacao_mais_baixa)
````

#i. Qual hor√°rio do dia cont√©m a maior m√©dia de loca√ß√µes de bicicletas?

````python
grupo_por_hora = df.groupby('hour')
media_por_hora = grupo_por_hora['total_count'].mean()
horario_mais_alto = media_por_hora.idxmax()
print("O hor√°rio que cont√©m a maior m√©dia de loca√ß√µes de bicicletas √©:", horario_mais_alto,"h")
````

#j. Qual hor√°rio do dia cont√©m a menor m√©dia de loca√ß√µes de bicicletas?

````python
grupo_por_hora = df.groupby('hour')
media_por_hora = grupo_por_hora['total_count'].mean()
horario_mais_baixo = media_por_hora.idxmin()
print("O hor√°rio que cont√©m a menor m√©dia de loca√ß√µes de bicicletas √©:", horario_mais_baixo,"h")
````

#k. Que dia da semana cont√©m a maior m√©dia de loca√ß√µes de bicicletas?


````python
df['datetime'] = pd.to_datetime(df['datetime'])
df['dia_semana'] = df['datetime'].dt.day_name()
grupo_por_dia_semana = df.groupby('dia_semana')
media_por_dia_semana = grupo_por_dia_semana['total_count'].mean()
dia_mais_alto = media_por_dia_semana.idxmax()
print("Dia da semana com a maior m√©dia de loca√ß√µes de bicicletas √©:", dia_mais_alto)
````

#l. Que dia da semana cont√©m a menor m√©dia de loca√ß√µes de bicicletas?

````python
df['datetime'] = pd.to_datetime(df['datetime'])
df['dia_semana'] = df['datetime'].dt.day_name()
grupo_por_dia_semana = df.groupby('dia_semana')
media_por_dia_semana = grupo_por_dia_semana['total_count'].mean()
dia_mais_baixo = media_por_dia_semana.idxmin()
print("Dia da semana com a menor m√©dia de loca√ß√µes de bicicletas √©:", dia_mais_baixo)
````

#m. √Äs quartas-feiras (weekday = 3), qual o hor√°rio do dia que cont√©m a maior m√©dia de loca√ß√µes de bicicletas?

````python
quartas_feiras = df[df['weekday'] == 3]
grupo_por_hora = quartas_feiras.groupby('hour')
media_por_hora = grupo_por_hora['total_count'].mean()
horario_mais_alto = media_por_hora.idxmax()
print("Hor√°rio do dia com a maior m√©dia de loca√ß√µes de bicicletas √†s quartas-feiras √©:", horario_mais_alto, "h")
````

#n. Aos s√°bados (weekday = 6), qual o hor√°rio do dia que cont√©m a maior m√©dia de loca√ß√µes de bicicletas?

````python
sabados = df[df['weekday'] == 6]
grupo_por_hora = sabados.groupby('hour')
media_por_hora = grupo_por_hora['total_count'].mean()
horario_mais_alto = media_por_hora.idxmax()
print("Hor√°rio do dia com a maior m√©dia de loca√ß√µes de bicicletas aos s√°bados √©:", horario_mais_alto, "h")
````
