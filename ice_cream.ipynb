```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/content/IceCreamData.csv')
```

```python
df
```

```python
result = df.sort_values('Temperature', ascending= False)
result
```

```python
plt.plot(result['Temperature'], result ['Sales'])
plt.title('Ice Cream Truck')
plt.show()
```

```python
plt.plot(result['Temperature'], result ['Sales'])
plt.title('Ice Cream Truck')
plt.ylabel('Sales (US$)')
plt.xlabel ('Temperature (Fahrenheit)')
plt.show()
```

```python
plt.plot(result['Temperature'], result ['Sales'], label = 'US$')
plt.title('Ice Cream Truck')
plt.ylabel('Sales (US$)')
plt.xlabel ('Temperature (Fahrenheit)')
plt.legend()
plt.show()
```

```python
plt.plot(result['Temperature'], result ['Sales'], 'r.--', label = 'US$')
plt.title('Ice Cream Truck')
plt.ylabel('Sales (US$)')
plt.xlabel ('Temperature (Fahrenheit)')
plt.legend()
plt.show()
```

```python
plt.bar(result['Temperature'], result['Sales'], color='blue', alpha=0.7, label='US$')
plt.title('Ice Cream Truck')
plt.ylabel('Sales (US$)')
plt.xlabel('Temperature (Fahrenheit)')
plt.legend()
plt.show()
```
